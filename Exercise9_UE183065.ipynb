**Read the Colleges dataset and clean all the attributes.**

1.   From all marks based attributes, remove % sign from different attributes
2.   In SAT/ACT, Keep higher value.
3.   From all cost based attributes, Remove $ sign and commas.
4.   Remove any null value in any attribute.


```python
import pandas as pd
import numpy as np

df = pd.read_csv("Top 100 Private Colleges.csv - Top 100 Private Colleges.2003.csv")
```

```python
df
```

```python
df2=df[['Admission Rate','*SAT or ACT','4-year Grad. Rate','6-year Grad. Rate','Aid From Grants','Non-Need-Based Aid+']].replace('\%','',regex=True).astype(str)
df[['Admission Rate','*SAT or ACT','4-year Grad. Rate','6-year Grad. Rate','Aid From Grants','Non-Need-Based Aid+']]=df2
df2=df['Need Met'].replace('\%','',regex=True).astype(str)
df2=df2.replace('NA','0',regex=True).astype(float)
df['Need Met']=df2
```

```python
df
```

```python
marks=df['*SAT or ACT'].tolist()
len(marks)
```

```python
high=[]
for i in marks:
  nums=i.split('/')
  num1=nums[0]
  if(num1[-1].isdigit()==False):
    num1=int(num1[:len(num1)-1])
  if(len(nums)>=2):
    num2=nums[1]
  else:
    num2='0*'      
  if(num2[-1].isdigit()==False):
    num2=int(num2[:len(num2)-1]) 
  try:   
    high.append(max(int(num1),int(num2)))
  except:
    pass  
```

```python
df['*SAT or ACT']=high
```

```python
df2=df[['Total Costs','Cost After Need-based Aid','Cost After Non-Need-Based Aid','Average Debt']].replace('\$','',regex=True).astype(str)
df2=df2[['Total Costs','Cost After Need-based Aid','Cost After Non-Need-Based Aid','Average Debt']].replace('\,','',regex=True).astype(str)
df[['Total Costs','Cost After Need-based Aid','Cost After Non-Need-Based Aid','Average Debt']]=df2
```

```python
df2=df['Non-Need-Based Aid+'].replace('nan',0)
```

```python
df['Non-Need-Based Aid+']=df2
```
